```python
cms2_mut_raw = pd.read_csv("/Users/violafanfani/Documents/uni-harvard/projects/tcga-data-supplement/data/processed/batch-coad-subtype-20250406/tcga_coad_cms2/data_download/mutations/tcga_coad_cms2_mutations.txt", sep=' ', index_col=0)
cms2_mut_raw
```

```python
cms2_mut_raw['subtype'] = 'cms2'
cms4_mut_raw['subtype'] = 'cms4'

mut_raw = pd.concat([cms2_mut_raw, cms4_mut_raw], axis=0)
```

```python
# Check average number of mutations per gene
mut_raw['Hugo_Symbol'].value_counts().sort_values(ascending=False).head(20)
```

```python
mut_raw[mut_raw['Hugo_Symbol'] == 'TP53'].head(10)
```

```python
sns.catplot(data=mut_raw[mut_raw['Hugo_Symbol'] == 'TP53'], y='Variant_Classification', hue='subtype', kind='count')
```

```python
mut_raw['nsamples'] = mut_raw['subtype'].map(mut_raw.groupby('subtype').nunique()['Tumor_Sample_Barcode'])
```

```python
# I want the counts to be rescaled by the number of samples
mut_raw['nmutations'] = mut_raw.groupby(['Hugo_Symbol', 'subtype'])['Tumor_Sample_Barcode'].transform('count')
mut_raw['nmutations'] = mut_raw['nmutations'] / mut_raw['nsamples']
mut_raw
```

```python
mut_raw['is_mutated'] = 1
mut_raw['perc_mutated'] = 1/mut_raw['nsamples']
```

```python
mut_raw[mut_raw['Hugo_Symbol'] == 'TP53'].groupby(['subtype', 'Variant_Classification']).agg({'is_mutated': 'sum', 'perc_mutated': 'sum'}).reset_index().sort_values('perc_mutated', ascending=False)
```

```python
# Plot nmutations
sns.barplot(data=mut_raw[mut_raw['Hugo_Symbol'] == 'TP53'].groupby(['subtype', 'Variant_Classification']).agg({'is_mutated': 'sum', 'perc_mutated': 'sum'}).reset_index().sort_values('perc_mutated', ascending=False), x='perc_mutated', y='Variant_Classification', hue='subtype', ci=None)
```
